```python
# Write a numpy program to implement following operation 
# to create a 3x3 matrix with values ranging from 2 to 10 

# Developed By: Solanki Tisha 
# Enrollement No: 216270307062

import numpy as np
x =  np.arange(2, 11)
y =  x.reshape(3,3)
print(y)
```

```python
# Write a numpy program to implement following operation 
# to append values at the end of an array 

# Developed By: Solanki Tisha 
# Enrollement No: 216270307062

import numpy as np
x = [9, 8, 7]
print("Original array:")
print(x)
x = np.append(x, [[6, 5, 4], [3, 2, 1]])
print("\nAfter append array:")
print(x)
```

```python
# Write a numpy program to implement following operation 
# to create another shape from an array without changing its data(3*2 to 2*3) 

# Developed By: Solanki Tisha 
# Enrollement No: 216270307062

import numpy as np
original = np.array([[11, 22],[33, 44],[55, 66]])


reshaped = original.reshape(2, 3)

print("3x2 Array:")
print(original)

print("\n2x3 Array:")
print(reshaped)
```
