```python
# Write a Numpy program to implement following operations 
# to split an array of 14 elements into 3 arrats each with 2,4, and 8 elements in the original order

#Developed By: Solanki Tisha
#Enrollment: 216270307062

import numpy as np 

array1 = np.arange(14)
print("Original Array",array1)
print("\n 2 Element Array",array1[:2])
print(" 4 Element Array",array1[2:6])
print(" 8 Element Array",array1[6:14])
```
