```python
# Write a numpy program to implement following operation 
# to compute the mean standard deviation and variance of a given array along the second axis 

# Developed By: Solanki Tisha
# Enrollement No: 216270307062

import numpy as np
data = np.array([[30, 20, 10],
                 [40, 60, 70],
                 [50, 80, 90]])

mean = np.mean(data, axis=1)
std_dev = np.std(data, axis=1)
variance = np.var(data, axis=1)

print("Mean along second axis:", mean)
print("Standard deviation along second axis:", std_dev)
print("Variance along second axis:", variance)
```
